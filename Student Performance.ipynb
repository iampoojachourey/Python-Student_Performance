```python
MathScore_List = list(dataset.MathScore)
WritingScore_List = list(dataset.WritingScore)
ReadingScore_List = list(dataset.ReadingScore)
```

```python
#Plot showing number of students who failed and passed the exam on the basis of a minimum pass marks of 35 
fail = 0
pass_stu = 0
count = 0
for count in MathScore_List:
    if (MathScore_List[count] < min_marks):
        fail = fail + 1
    elif (WritingScore_List[count] < min_marks):
        fail = fail + 1
    elif(ReadingScore_List[count] < min_marks):
        fail = fail + 1
    else:
        pass_stu = pass_stu+1
print("Number of students who failed the exam:", fail)
print("Number of students who passed the exam:", pass_stu)

result = (fail, pass_stu)
plot1 = sns.barplot(data=dataset, x=Result, y= Result)
```
